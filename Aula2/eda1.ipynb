```python
import numpy as np
import pandas as pd
import seaborn as sns
from scipy import stats
import matplotlib.pyplot as plt
```

### Média, Mediana e Moda

```python
lista_numeros = np.array([10, 20, 20, 30, 40, 50, 60, 100, 250, 400, 600])
print(f"Média: {lista_numeros.mean():.2f}\nMediana: {np.median(lista_numeros):.2f}\nModa: {stats.mode(lista_numeros)[0]:.2f}")
```

### Quantis, Quartis, Percentis e Mediana.

```python
print(f"Valor do 50º percentil: {stats.scoreatpercentile(lista_numeros, 50)}.")
print(f"Percentil correspondente ao valor 3800: {stats.percentileofscore(lista_numeros, 3800)}.")
```

### Histogramas

```python
sns.set_style("whitegrid")
ax = sns.histplot(lista_numeros, bins=3)
ax.set(xlabel="Intervalos", ylabel="Quantidades")
plt.show()
```

### Boxplots

```python
sns.set_style("whitegrid")
ax = sns.boxplot(lista_numeros)
ax.set(ylim=(0, 700), ylabel="Valores")
plt.show()
```

```python
# Create a KDE plot using Seaborn
sns.set(style="whitegrid")
sns.kdeplot(lista_numeros, fill=True)
sns.scatterplot(x=lista_numeros, y=[0.00005]*len(lista_numeros), color='red', marker='o', s=50)

# Add labels and title
plt.xlabel('Valores')
plt.ylabel('Densidade')

# Show the plot
plt.show()
```

### Momento

- k = 1: 0;
- k = 2: 1;
- k = 3: *skewness*;
- k = 4: *kurtosis*.

```python
k = 3 
n = len(lista_numeros)
media = lista_numeros.mean()
std_lista_numeros = lista_numeros.std()
diff_lista_numeros = [(xi - media)**k for xi in lista_numeros]
print(sum(diff_lista_numeros) / (n*(std_lista_numeros**k)))
```
