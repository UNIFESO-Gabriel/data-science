```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.pyplot import figure
from scipy.cluster.hierarchy import dendrogram, linkage
from sklearn.preprocessing import StandardScaler, MinMaxScaler, LabelEncoder
```

```python
df_pokemon = pd.read_csv("../dados/pokemon_anderson/pokemon.csv", sep=',')
df_pokemon.columns
```

```python
# Clusterizando por peso e velocidade
x, y = df_pokemon['weight_kg'], df_pokemon['speed'],
data = list(zip(x, y))

# Atribuindo rótulos
str_names = [nome for nome in df_pokemon['name'].values]
```

```python
# Plota os pontos
plt.scatter(x[:10], y[:10])

# Plota os rótulos
for i, txt in enumerate(df_pokemon["name"][:10]):
    plt.annotate(txt, (x[i], y[i]))

#fig, ax = plt.subplots()
plt.show()
```

```python
# Para visualização
from matplotlib.pyplot import figure

# Para a clusterização
from scipy.cluster.hierarchy import dendrogram, linkage

figure(figsize=(18, 6), dpi=80)
linkage_data = linkage(data[:10], method='ward', metric='euclidean')
dendrogram(linkage_data, labels=str_names[:10])

plt.show()
```
