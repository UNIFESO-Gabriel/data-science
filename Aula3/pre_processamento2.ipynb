```python
import pandas as pd
import numpy as np
from sklearn.decomposition import PCA
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler, StandardScaler
from sklearn.metrics import confusion_matrix, classification_report, mean_squared_error
```

```python
df_games_sales = pd.read_csv("../dados/game-sales-kaggle/game_sales.csv", header=0)

cat_column_names = ["Platform", "Genre", "Publisher"]
df_games_sales[cat_column_names] = df_games_sales[cat_column_names].astype("category")

for column_name in cat_column_names:
    df_games_sales[column_name] = df_games_sales[column_name].cat.codes
    
X_train, X_test, y_train, y_test = train_test_split(df_games_sales, df_games_sales["Global_Sales"], test_size=0.2)
```

```python
df_games_sales["Year"] = df_games_sales["Year"].fillna(df_games_sales["Year"].mean()).astype(int)
df_games_sales["Rank"] = df_games_sales["Rank"].astype(int)
df_games_sales
```

```python
scaler = StandardScaler()
X_test["Year"] = X_test["Year"].fillna(X_test["Year"].mean()).astype(int)
arr = scaler.fit_transform(X_test.drop(["Name"], axis=1))
print(arr.shape)

pca = PCA(n_components=5)
arr = pca.fit_transform(arr)
print(arr.shape)
```

```python
clf = LinearRegression()

X_train["Year"] = X_train["Year"].fillna(X_train["Year"].mean()).astype(int)
clf.fit(X_train.drop("Name", axis=1), X_train["Global_Sales"])

X_test["Year"] = X_test["Year"].fillna(X_test["Year"].mean()).astype(int)
y_pred = scaler.fit_transform(X_test.drop(["Name"], axis=1))
y_pred = pca.fit_transform(y_pred)

y_pred = clf.predict(X_test.drop("Name", axis=1))

print(mean_squared_error(X_test["Global_Sales"], y_pred))
```
